```python
def extract_pdf_text(file_path):
    pages = []
    
    with pdfplumber.open(file_path) as pdf:
        for i, page in enumerate(pdf.pages):
            text = page.extract_text()
            pages.append({
                "page_number": i + 1,
                "text": text
            })
    
    return pages

extract_text=extract_pdf_text(file_path='immigration_act_2025.pdf')
```

```python
print(extract_text[0]['text'])
```

```python
def clean_text(text):
    if not text:
        return ""
    
    text = text.replace("\n", " ")   # remove line breaks
    text = " ".join(text.split())   # remove extra spaces
    
    return text
```

```python
def process_pdf(pages):
    cleaned_pages=[]
    
    
    for page in pages:
        cleaned=clean_text(page['text'])
        
        cleaned_pages.append({
            'page': page['page_number'],
            'content':cleaned
        })
    return cleaned_pages

process_pdf(pages=extract_text)
        
        
```

```python
#Chunk Each Page (Sentence + Overlap)
import re

def chunk_with_overlapping_sentence(text, max_sentences, overlap):
    sentences=re.split(r'(?<=[.!?])[\s-]+', text)
    chunks=[]
    
    Step= max_sentences-overlap
    for i in range(0, len(sentences), Step):
        chunk=" ".join(sentences[i:i+max_sentences])
        chunks.append(chunk)
    return chunks
```

```python
#Combine Everything (FULL PIPELINE)
def process_pdf_with_chunk(pages):
    all_chunks =[]
    
    
    for page in pages:
        cleaned=clean_text(page['text'])
        chunks = chunk_with_overlapping_sentence(cleaned, max_sentences=10, overlap=1)
        all_chunks .append({
            'page': page['page_number'],
            'content':chunks
        })
    return all_chunks 

result= process_pdf(pages=extract_text)
```

```python
for i, c in enumerate(result):
     print(f"Chunk {i+1}: {c}")
```
