```python
import re
```

```python
text = "Machine learning is a field of AI. It allows systems to learn from data."
text_from_file= open('sample_text.txt', 'r', encoding='utf-8')
text_from_file=text_from_file.read()
```

```python

#solution Split text every N characters
def chunk_by_N_character(text, chunk_size):
    chunks=[]
    for i in range( 0, len(text), chunk_size):
        chunk=text[i:i+chunk_size]
        chunks.append(chunk)
    print(chunks)

chunk_by_N_character(text=text, chunk_size=50)
```

**Problem chunking by charecter**<br>
- Words are cut ❌
- Sentences broken ❌
- Meaning lost ❌

 This is not good for RAG

```python
#solution word based chunking

def chunking_by_words(text, chunk_size):
    words= text.split()
    chunks=[]
    
    for i in range(0, len(words), chunk_size):
        chunk= " ".join(words[i:i+chunk_size])
        chunks.append(chunk)
        print(f"chunk {i+1} : {chunk}")
    
chunking_by_words(text=text, chunk_size=5)
```

**Problems word based chunking :**<br>
- Sentence may still break ❌
- Meaning may still split ❌

```python
#solution : sentence chunking 

def chunking_by_sentence(text, max_sentences):
    sentences= re.split(r'(?<=[.!?]) +', text)
    chunks=[]
    
    for i in range(0, len(sentences), max_sentences):
        chunk=" ".join(sentences[i:i + max_sentences])
        chunks.append(chunk)
    
    return chunks
        
result = chunking_by_sentence(text=text, max_sentences=1)
for i, c in enumerate(result):
     print(f"Chunk {i+1}: {c}") 
```

**problem of sentence chunking** <br>
- it looses context
- meaning become weeks


```python
#solution: overlapping chunking
def chunk_with_overlapping_sentence(text, max_sentences, overlap):
    sentences=re.split(r'(?<=[.!?])[\s-]+', text)
    chunks=[]
    
    Step= max_sentences-overlap
    for i in range(0, len(sentences), Step):
        chunk=" ".join(sentences[i:i+max_sentences])
        chunks.append(chunk)
    
    return chunks
result= chunk_with_overlapping_sentence(text=text_from_file, max_sentences=2, overlap=1)

for i, c in enumerate(result):
     print(f"Chunk {i+1}: {c}")
```

**Trade-Off of Overlap**
- More chunks → more storage ❌
- More computation ❌
- Slight redundancy ❌

👉 But:
✅ Worth it for better RAG quality
